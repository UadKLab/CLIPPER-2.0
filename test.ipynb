```python
# a function that reads a fasta file with Biopython and saves records in a dictionary
from Bio import SeqIO

def read_fasta(fasta_file):
    """Reads a fasta file and returns a dictionary with record.id as keys and record.seq as values."""
    fasta_dict = {}
    for record in SeqIO.parse(fasta_file, "fasta"):
        fasta_dict[record.id] = record.seq
    return fasta_dict
```
